```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes_X = diabetes.data[:, np.newaxis, 2]
```

```python
diabetes_X_train = diabetes_X[:-20] #all data except last 20 entries
diabetes_X_test = diabetes_X[-20:] #selects last 20 elements/entries
```

```python
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(diabetes_X_train, diabetes_y_train)
diabetes_y_predicted = model.predict(diabetes_X_test)
```

```python
print('Coefficients: \n', model.coef_)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_predicted))
```

```python
print('Variance score: %.2f' % r2_score(diabetes_y_test, diabetes_y_predicted))
```

```python
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
```

```python
plt.plot(diabetes_X_test, diabetes_y_predicted, color='blue', linewidth=3)
```
